### Part 1. _A critical analysis and redesign towards Nightingale's coxcomb_

Data visualization is now well established as an integral part of any scientific report and presentation. Looking back in history, Florence Nightingale made her most significant advances in data visualization when war and public health were essential partners in improving health care. In this analysis, a redesign of the famous Nightingaleâ€™s rose chart is conducted using a modern tool with code libraries (python with pyechart library). The re-structure is based on three key steps: analyzing the original work, critiquing the weakness, and proposing new ideas.

The famous rose diagram is first published in Notes on _Matters Affecting the Health, Efficiency, and Hospital Administration of the British Army. Founded Chiefly on the Late War Experience_ by a nurse, statistician, and reformer Florence Nightingale in 1858. The original work, "The cause of mortality in the army in the East," illustrated the seasonal mortality rates of soldiers in the Crimean War between 1854 and 1856, as shown in *Figure 1*. Nightingale used a visual format to help the decision-makers at the time understand complex statistics in an easy, distinct, and convincing way. To be specific, the blue, red, and black wedges represent three kinds of causes of soldiers' deaths, while each fan shape represents the number of deaths in each month. The larger the area, the higher the death rate. From the figure, one could directly determine the most influential factors by finding the largest sector area. The coxcomb reads the cyclic data clockwise, and people can know the trend of data and the proportion of subcategories over time. Comparing the right "not improved" group before April 1855 with the left "improved group" after that time, the figure points out that most soldiers die not on the battlefield but in poor medical conditions. This visual aid has prompted the army to re-examine how soldiers are treated and reformed the way the army operates.

Nightingale's rose diagram is meaningful but not flawless, and two main drawbacks occur in it. **First, one of the limitations is that the rose diagram exaggerates the data's scale visually, which would lead to people's false conclusion.** Precisely, the diagram reflects the value in terms of radius. However, the relationship between radius and area is squared, which means people will be misled and think that the existing differences are more dramatic than they imagined. The more noticeable outer segments are inversely proportional to the increment of the number of people who died due to diseases, resulting in misunderstanding. Even though the fan area's exaggerated effect helps compare similar values, the rose diagram is not necessarily a good choice to represent accurate results. **The second disadvantage of the diagram is that it is hard for people to compare the effect among different factor groups directly (i.e., the number of deaths due to diseases, wounds, or other reasons) and the impact before and after the treatment (i.e., renovation).** One possible reason is that people intuitively think the color blocks are stacked together rather than overlapping. For instance, it is hard to compare the impact of the number of deaths between diseases and wounds in the same month as the areas of those two factors are not in the same horizontal line. Another example will be the red areas before and after the 1855 renovation look nearly the same. People will draw an unrigorous conclusion based on the diagram that the improved medical care does not affect wounded soldiers. Aware that any data visualization's priority should be to ensure the story's accuracy, including clarity of content in the data age. The rose diagram has the advantage of making a strong visual impression on decision-makers, but the exaggerated effect can make incorrect conclusions. 


_A grouped (or column) bar chart with dynamic effects_ provides a new facet to visualize Nightingale's dataset to address these two concerns mentioned previously. The redesign plot is visualized using a powerful visualization package called PyEcharts through Jupyter Notebook. **This plot is based on a traditional grouped bar plot and extends interaction effects, which help people reason data more compellingly.** Indeed, the grouped bar plot is similar to the rose diagram but merely retains its functionality. In particular, rectangular bars are used to indicate the number of deaths in each month. For instance, *Figure 2* shows that it allows people to compare each group's difference without exaggerated effects accurately. **Besides, it also well-addressed the second issue in the original diagram with dynamic visualization since the redesign includes two new features: different groups and a specific range selection.** **The first feature allows for within-group comparisons;** that is, one can figure out how the cause of death changes in each month. As shown in *Figure 3*, the dynamic redesign plot utilizes the same color palette as Nightingale's coxcomb and can instantly display data for different groups based on selection. As the selection changes, the data points' scale also changes, which could avoid misleading due to the data's difference ratio. **On the other hand, the second feature uses the DataZoom component to select a specific range for between-groups comparison.** Note that one of Nightingale's rose diagram's key advantages is that it divided the graph into two years so that people could compare the effect before and after the change of sanitary condition. Professor Hugh Small of the University of Durham in England published the article "Did Nightingale's 'Rose Diagram' or 'Coxcomb' Save Millions of Lives?" in 2010, and pointed out that pulling all the data into a single bar plot will reduce comparison over years and seasons. Indeed, the season is a possible factor affecting the number of deaths, but the plot's goal is to indicate the impact of diseases. In this case, this analysis uses the dynamic timeline feature from the pyecharts library to address the concerns. As reflected in *Figure 4*, the restructured plot demonstrates the overall trend of the dataset and allows people to magnify a certain period and compare those results. In short, compared to the original rose chart, a grouped bar chart with a dynamic interaction effect exists as a more efficient, interactive, and better-looking alternative with features of specific group and range comparison.

```python
# import bar plot
from pyecharts import Bar
```

```python
# Data retrived from the R HistData package: https://rdrr.io/cran/HistData/man/HistData-package.html
df = pd.read_csv("nightingale.csv")
```

```python
df.head()
```

#### 2: Select required variables 

```python
#select the column, and convert it to a list
Date=df["Date"].values.tolist()
Disease=df["Disease"].values.tolist()
Wounds = df["Wounds"].values.tolist()
Other = df["Other"].values.tolist()
```

#### 3: Redesign

#### Version 1 (basic)

Given the similar functionality of the rose chart and bar charts, the reset will be based on the bar plot. <br/>
The basic one is the original bar charts which divide the cause of death into three groups: disease, wonds, and other.

```python
#used bar function
redesign0 = Bar("Redesign: Cause of Death from 1854-04 to 1856-03")
redesign0.add("Disease", Date, Disease,
             # did not perform a stacked bar plot, default is False
             is_stack=False,label_color=["#1d2228"]) 

redesign0.add("Wounds", Date, Wounds, 
             is_stack=False,label_color=["#bd4545"])

redesign0.add("Other", Date, Other, 
             is_stack=False,label_color=["#66859f"],legend_pos="right")
```

```python
redesign0
```

#### Version 2 (add-on features)

Note that the basic plot did not indicate the particular significant number (such as min, max, etc.). It is hard to compare between groups directly. Also, as mentioned previously, pulling all the data into a single bar plot will reduce comparison over years and seasons, which is the opposite of what it was redesigned to be. Therefore, to make people compare each group's difference accurately, the average lines and max points are added. The second version of the plot is shown as follows.

```python
#used bar function
redesign1 = Bar("Redesign: Cause of Death from 1854-04 to 1856-03")
redesign1.add("Disease", Date, Disease,
             # did not perform a stacked bar plot, default is False
             is_stack=False,label_color=["#1d2228"],
             #add a markline to indicate the average, and a point to indicate the max
            mark_line=["average"],mark_point=["max","min"]) 

redesign1.add("Wounds", Date, Wounds, 
             is_stack=False,label_color=["#bd4545"],
             mark_line=["average"],mark_point=["max"])

redesign1.add("Other", Date, Other, 
             is_stack=False,label_color=["#66859f"],
             mark_line=["average"],mark_point=["max"],
             #put the legend to the top right corner
             legend_top="top",legend_pos="right") 
```

```python
redesign1
```

#### Final Version

It is worth noticing that one of the critical features of Nightingale's rose chart is that she used two coxcombs to compare the data before and after changing medical conditions. Therefore, a data zoom feature is added to retain this feature, allowing people to select and compare specific time ranges. 

```python
#used bar function
redesign = Bar("Redesign: Cause of Death from 1854-04 to 1856-03")
redesign.add("Disease", Date, Disease,
             # did not perform a stacked bar plot, default is False
             is_stack=False,label_color=["#1d2228"],
             #add a markline to indicate the average, and a point to indicate the max
            mark_line=["average"],mark_point=["max","min"]) 

redesign.add("Wounds", Date, Wounds, 
             is_stack=False,label_color=["#bd4545"],
             mark_line=["average"],mark_point=["max"])

redesign.add("Other", Date, Other, 
             is_stack=False,label_color=["#66859f"],
             mark_line=["average"],mark_point=["max"],
             #put the legend to the top right corner
             legend_top="top",legend_pos="right",
             # add zoom-in and zoom-out component to the x-axis so that it is possible to magnify a small area
             datazoom_type="slider",is_datazoom_show=True) 
```

```python
redesign
```
